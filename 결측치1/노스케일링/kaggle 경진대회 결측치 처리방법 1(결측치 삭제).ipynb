### 1. 전체 데이터 중 결측치 있는 항목(행)을 삭제합니다.
 - 결측치 있는 컬럼 : workclass, occupation, native-country

```python
# rem_MV(remove missing value)에 workclass가 " ?"인 모든 행을 저장합니다.
# drop함수 이용해서 해당 행 삭제합니다.
rem_MV = train[train['workclass'] == " ?"].index
train = train.drop(rem_MV)
```

```python
rem_MVt = test[test['workclass'] == " ?"].index
test = test.drop(rem_MVt)
```

```python
# 같은 방식으로 occupation이 "?"인 모든 행을 삭제합니다
rem_MV1 = train[train['occupation'] == " ?"].index
train = train.drop(rem_MV1)
```

```python
rem_MVt1 = test[test['occupation'] == " ?"].index
test = test.drop(rem_MVt1)
```

```python
rem_MV2 = train[train['native-country'] == " ?"].index
train = train.drop(rem_MV2)
```

```python
rem_MVt2 = test[test['native-country'] == " ?"].index
test = test.drop(rem_MVt2)
```
