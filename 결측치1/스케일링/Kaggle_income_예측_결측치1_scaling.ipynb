### 결측치 채우기 1번 방법 : 결측치 제거

- workclass, occupation, native-country의 결측치 행 제거

```python
# rem_MV(remove missing value)에 workclass가 " ?"인 모든 행을 저장합니다.
# drop함수 이용해서 해당 행 삭제합니다.
rem_MV = train[train['workclass'] == " ?"].index
train = train.drop(rem_MV)
```

```python
train[train['workclass'] == " ?"]
```

```python
rem_MVt = test[test['workclass'] == " ?"].index
test = test.drop(rem_MVt)
```

```python
test[test['workclass'] == " ?"]
```

```python
# 같은 방식으로 occupation이 "?"인 모든 행을 삭제합니다
rem_MV1 = train[train['occupation'] == " ?"].index
train = train.drop(rem_MV1)
```

```python
train[train['occupation'] == " ?"]
```

```python
rem_MVt1 = test[test['occupation'] == " ?"].index
test = test.drop(rem_MVt1)
```

```python
test[test['occupation'] == " ?"]
```

```python
rem_MV2 = train[train['native-country'] == " ?"].index
train = train.drop(rem_MV2)
```

### 문제, 정답 데이터로 분리

```python
# train 데이터에서 정답 컬럼
y_train = train['income']
# train 데이터에서 정답 컬럼을 삭제한 후 남은 모든 컬럼들
X_train = train.drop('income', axis=1)
# 맨 처음에 받았던 test.csv
X_test = test
```

### 데이터 인코딩 하기

```python
X_train = pd.get_dummies(X_train)
X_test = pd.get_dummies(X_test)
# X_train
# X_test
```

```python
print(X_train.shape)
print(X_test.shape)
```

### 컬럼 개수 맞추기

```python
set(X_train.columns) - set(X_test.columns)
```

```python
# 'native-country_ Holand-Netherlands'의 컬럼은 생기지만 실제로 컴퓨터가 데이터는 없는 것으로 인식함!
X_test['native-country_ Holand-Netherlands'] = 0
```

```python
tmp = X_train.columns
for x in tmp :
    print(x)
```

```python
tmp = X_test.columns
for x in tmp :
    print(x)
```

```python
# 위치 변경을 위해 'native-country_ Holand-Netherlands' 컬럼을 tmp 변수에 넣어줌

tmp = X_train['native-country_ Holand-Netherlands']
```

```python
X_train.drop('native-country_ Holand-Netherlands', axis = 1, inplace=True)
```

```python
X_train = pd.concat([X_train, tmp],axis=1)
X_train
```

### Scaler 만들기

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(X_train)
```

```python
transform_X_train = scaler.transform(X_train)
transform_X_train
```

```python
transform_X_test = scaler.transform(X_test)
transform_X_test
```

```python
from sklearn.model_selection import cross_val_score, train_test_split, KFold
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
```

```python
from lightgbm import LGBMClassifier 

lgb_model = LGBMClassifier(n_estimators = 1000, learning_rate=0.01, random_state=123)
lgb_model.fit(transform_X_train, y_train)
```

```python
from xgboost import XGBClassifier

xgb_model = XGBClassifier(n_estimators = 100, learning_rate=0.01, random_state = 123)
xgb_model.fit(transform_X_train, y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier

forest_model = RandomForestClassifier(n_estimators=100, random_state= 11)
forest_model.fit(transform_X_train, y_train)
```

### 정확도 분석하기

kfold 교차 검증

```python
# ---------------------------------------------

# 정확도

kfold = KFold(n_splits=10, random_state=123, shuffle=True)

acc_result = cross_val_score(lgb_model, # 모델
                         X_train, # transform_X_train
                         y_train,
                         cv = kfold,
                        )

print(acc_result.mean())

# -----------------------------------------------

# F1 score

f1_result = cross_val_score(lgb_model, # 모델
                         X_train, # transform_X_train
                         y_train,
                         cv = kfold,
                         scoring='f1'
                        )

print(f1_result.mean())

# ------------------------------------------
```

```python
# ---------------------------------------------

# 정확도

kfold = KFold(n_splits=10, random_state=123, shuffle=True)

acc_result = cross_val_score(xgb_model, # 모델
                         X_train, # transform_X_train
                         y_train,
                         cv = kfold,
                        )

print(acc_result.mean())

# -----------------------------------------------

# F1 score

f1_result = cross_val_score(xgb_model, # 모델
                         X_train, # transform_X_train
                         y_train,
                         cv = kfold,
                         scoring='f1'
                        )

print(f1_result.mean())

# ------------------------------------------
```

```python
# ---------------------------------------------

# 정확도

kfold = KFold(n_splits=10, random_state=123, shuffle=True)

acc_result = cross_val_score(forest_model, # 모델
                         X_train, # transform_X_train
                         y_train,
                         cv = kfold,
                        )

print(acc_result.mean())

# -----------------------------------------------

# F1 score

f1_result = cross_val_score(forest_model, # 모델
                         X_train, # transform_X_train
                         y_train,
                         cv = kfold,
                         scoring='f1'
                        )

print(f1_result.mean())

# ------------------------------------------
```
